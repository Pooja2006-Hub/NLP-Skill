```python
# Download required NLTK data packages
import nltk
nltk.download('maxent_ne_chunker_tab')
nltk.download('words')  # Also needed for named entity recognition
nltk.download('punkt')  # For tokenization
nltk.download('averaged_perceptron_tagger')  # For POS tagging

# Import required modules
from nltk import word_tokenize, pos_tag, ne_chunk
from nltk.tree import Tree
from nltk.classify import NaiveBayesClassifier
from nltk import FreqDist, bigrams, trigrams

# Your original code starts here
# Note: You'll need to define the 'news' variable first
news = "Your news text here"  # Add your actual news text

tokens = word_tokenize(news)
tagged = pos_tag(tokens)
entities = ne_chunk(tagged)

for subtree in entities:
    if isinstance(subtree, Tree):
        entity_name = " ".join([token for token, pos in subtree.leaves()])
        print(subtree.label() + ":", entity_name)

# -------------------------------
# 3. Text Classification
# -------------------------------
def features(text):
    words = text.lower().split()
    return {word: True for word in words}

training_data = [
    (features("football cricket match tournament"), "Sports"),
    (features("government election parliament minister"), "Politics"),
    (features("market stock finance business"), "Business"),
    (features("AI software technology platform"), "Technology")
]

classifier = NaiveBayesClassifier.train(training_data)

category = classifier.classify(features(news))
print("\nCategory:")
print(category)

# -------------------------------
# 4. Model Evaluation
# -------------------------------
test_set = training_data
accuracy = nltk.classify.accuracy(classifier, test_set)

print("\nModel Accuracy:")
print(f"{accuracy*100:.0f}%")

# -------------------------------
# 5. Word Frequency & Probability
# -------------------------------
words = [w.lower() for w in word_tokenize(news)
         if w.isalpha()]

fdist = FreqDist(words)

print("\nWord Frequency:")
for word, freq in fdist.most_common():
    print(word, ":", freq)

print("\nProbability Distribution:")
total_words = len(words)

for word, freq in fdist.items():
    print(f"P({word}) = {freq}/{total_words} = {freq/total_words:.2f}")

# -------------------------------
# 6. Semantic Analysis
# -------------------------------
print("\nSemantic Relationships:")
print("Microsoft -> AI Platform")
print("AI -> Technology")
print("Technology -> Business Productivity")

# -------------------------------
# 7. Bigrams and Trigrams
# -------------------------------
print("\nBigrams:")
for bg in bigrams(words):
    print(bg)

print("\nTrigrams:")
for tg in trigrams(words):
    print(tg)
```
