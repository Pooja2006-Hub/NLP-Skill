```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.probability import FreqDist

# Download required resources (run once) - Updated to include punkt_tab
nltk.download('punkt_tab')  # Changed from 'punkt' to 'punkt_tab'
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger_eng')  # Changed to language-specific version

# Sample Resume Text
resume = """
John is a Computer Science graduate with experience in Python programming,
Machine Learning, Data Analysis, and Web Development. He has worked on
AI projects and developed software applications.
"""

# 1. Tokenization
tokens = word_tokenize(resume)
print("1. Tokens:")
print(tokens)

# 2. Stopword Removal
stop_words = set(stopwords.words('english'))
filtered_words = [word.lower() for word in tokens
                  if word.isalpha() and word.lower() not in stop_words]

print("\n2. After Stopword Removal:")
print(filtered_words)

# 3. Stemming
stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in filtered_words]

print("\n3. Stemmed Words:")
print(stemmed_words)

# 4. Lemmatization
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_words]

print("\n4. Lemmatized Words:")
print(lemmatized_words)

# 5. POS Tagging
pos_tags = nltk.pos_tag(filtered_words)

print("\n5. POS Tags:")
print(pos_tags)

# 6. Frequency Distribution
freq_dist = FreqDist(filtered_words)

print("\n6. Frequency Distribution:")
for word, freq in freq_dist.items():
    print(word, ":", freq)
```
